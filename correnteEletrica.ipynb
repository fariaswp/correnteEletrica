```python
########################################################################
# Defini bibliotecas a serem utilizadas

import numpy as np              #Definindo biblioteca de calculos numericos
import matplotlib.pyplot as plt #Definindo biblioteca de graficos
import math                     #Definindo biblioteca de constantes
```

```python
########################################################################
# Caracterização de Ip: corrente do primário ou condutor

# Inicialização dos Parametros
Ap=6.4                        # amplitude do sinal senoidal
f=60                          # frequencia fundamental da rede
dfRede =0                     # representa defasagem rede eletrica em relação valor experimental medido
ph = (dfRede)*(math.pi/360)   # fase
Nciclos=3                     # número de ciclos de corrente desejado
tf = Nciclos*(1/f)            # tempo final desejado
fs = 40*1000                     # frequencia de amostragem
t= np.arange(0, tf, (1/fs))      # vetor de tempo de 0 a tf com intervalos de 1/fs t=0:1/fs:tf
ip= Ap*np.sin(2*math.pi*f*t+ph)  # corrente no primário
```

```python
########################################################################
# Definindo representação grafico
# fig : instancia da figura = quadro branco
# ax  : e um frama = area para plotar

fig , ax = plt.subplots()
ax.plot(t,ip,'b-', linewidth=1)
ax.set_title('Corrente eletrica alternada')

plt.xlabel('Tempo(s)')
plt.ylabel('Ip (A)')
plt.show()
```
